```python
#how to read all files#

#os.listdir(path)是得到在path路径下所以文件的名称列表.
#open(path)是打开某个文件.
#iter是python的迭代器.
import os

path='/Users/huatongyan/Desktop/excel_processing/risk200917'
#path=os.getcwd()


def check_file(file_path):
    os.chdir(file_path)
    print(os.path.abspath(os.curdir))
    all_file = os.listdir()
    files = []
    for f in all_file:
        if os.path.isdir(f):
            a=check_file(file_path+'/'+f)
            p=os.getcwd()
            for ff in a:
                #print(str(p+ff))
                files.append(p+'/'+ff)
            #print(p,a)
            #files.extend(check_file(file_path+'/'+f))
            os.chdir(file_path)
        else:
            files.append(f)
    return files
 
file_list = check_file(path)   

import re
ds=[]
for i in range(len(file_list)):
    a=re.search(r'.DS_Store',file_list[i])
    if a!=None:
        ds.append(file_list[i])
for j in ds:
    file_list.remove(j)
#file_list
```

```python
#读取子表功能实现#

import openpyxl
import pandas as pd
import numpy as np

def get_row(filename):
    df=pd.read_excel(filename,sheet_name='01-风险管理表模板')
    name=df.iloc[0][7].replace('项目名称：','')
    df['项目名称']=name
    df=df.iloc[5:,[28,5,3,4,9,10,11,15,16]]

    df.rename(columns={
                   'Unnamed: 5':'风险描述',
                   'Unnamed: 3':'一级类别',
                   'Unnamed: 4':'二级类别',
                   'Unnamed: 9':'风险评分',
                   'Unnamed: 10':'响应策略',
                   'Unnamed: 11':'需要协调的资源',
                   'Unnamed: 15':'响应后风险评分',
                   'Unnamed: 16':'现状',
                  }, inplace = True)
    df.dropna(subset=['风险描述'], how='all', inplace=True)
    return df
 
def concat_df():
    filename1=file_list[0]
    df=get_row(filename1)
    for filename in file_list:
        try:
            df=pd.concat([df,get_row(filename)],axis=0)
        except:
            print(filename,' has some bug')
    return df
concat_df()
```
