```python
#dependency：xlrd
#工作目录为文件夹
import openpyxl
import pandas as pd
import numpy as np

def clean_transform(filename,csvname):
    workbook1 = openpyxl.load_workbook(filename)#最后记得workbook1.close()
    sheet1=workbook1.active
    #1.取消合并单元格
    m_list = sheet1.merged_cells
    cr = []
    cr_n=[]
    for m_area in m_list:
        # 合并单元格的起始行坐标、终止行坐标。。。。，
        r1, r2, c1, c2 = m_area.min_row, m_area.max_row, m_area.min_col, m_area.max_col
        # 纵向合并单元格的位置信息提取出
        if r2 - r1 > 0:
            cr.append((r1, r2, c1, c2))
            #print('符合条件%s' % str(m_area))
    for r in cr:
        sheet1.unmerge_cells(start_row=r[0],start_column=r[2],end_row=r[1],end_column=r[3])
        for row in range(r[0], r[1]):
            sheet1.cell(row=row+1, column=r[3], value=sheet1.cell(r[0], r[3]).value)
    #2.删除表头（删除保留到行名称）
    tryind=['A1','A2','A3','A4','A5','A6','A7']
    for i in tryind:
        if sheet1[i].value=='序号':
            for j in range(1,tryind.index(i),1):
                sheet1.delete_rows(j)
            break
    workbook1.save('filename')
    workbook1=pd.read_excel('filename',header = 1)
    workbook1.to_csv('csvname', encoding='utf-8')
    workbook1.columns = workbook1.columns.map(lambda x:str(x).replace('账号',''))
    workbook1.columns = workbook1.columns.map(lambda x:str(x).replace('bip','BIP'))
    workbook1['项目组']=workbook1['项目组'].str.replace('\n','')
    workbook1['工作工时']=workbook1.where(workbook1.applymap(type).eq(float))[['工作工时']]
    #workbook1['工作工时']=workbook1['工作工时'].map(lambda x:float(str(x).replace('\\','0')))
    workbook1.fillna(0,inplace=True)
    #workbook1['产品线']=workbook1['产品线'].apply(lambda x:str(x).replace('\n',''))
    return workbook1

def sum_processing(df1):
    df1=df1.fillna(0)
    df_mul=df1.groupby('BIP')[['工作工时']].count()
    df_mul=df_mul[~(df_mul==0).all(axis=1)]
    df_mul=df_mul[~(df_mul==1).all(axis=1)]
    person_jianzhi=df_mul.index.tolist()
    df_mul.rename(columns={'工作工时':'兼职组数'},inplace=True)
    person_jianzhi=df_mul.index.tolist()
    
    df_time_sum=df1[df1['BIP'].isin(person_jianzhi)].groupby('BIP')[['工作工时']].sum()
    df_mul_time_sum=pd.merge(df_mul,df_time_sum,how='outer',on='BIP')
    ddff1=pd.merge(df1,df_mul_time_sum,how='left',on='BIP')
    ddff1['人头数']=ddff1['工作工时_x']/ddff1['工作工时_y']
    ddff1['人头数'].fillna(1,inplace=True)
    df_new_sum=ddff1.groupby('项目组')[['工作工时_x','人头数']].sum()
    df_new_sum['日组均工时']= df_new_sum['工作工时_x']/df_new_sum['人头数']
    return df_new_sum
def getday_table(xlsx1,csv1,xlsx2,csv2,xlsx3,csv3,xlsx4,csv4,xlsx5,csv5,
                 xlsx6,csv6,xlsx7,csv7,xlsx8,csv8,xlsx9,csv9,xlsx10,csv10,
                 xlsx11,csv11,xlsx12,csv12,xlsx13,csv13,
                 work_h,work_avgh,csvname):
    #202006013
    df1=clean_transform(xlsx1,csv1)
    df2=clean_transform(xlsx2,csv2)
    df3=clean_transform(xlsx3,csv3)
    df4=clean_transform(xlsx4,csv4)
    df5=clean_transform(xlsx5,csv5)
    df6=clean_transform(xlsx6,csv6)
    df7=clean_transform(xlsx7,csv7)
    df8=clean_transform(xlsx8,csv8)
    df9=clean_transform(xlsx9,csv9)
    df10=clean_transform(xlsx10,csv10)
    df11=clean_transform(xlsx11,csv11)
    df12=clean_transform(xlsx12,csv12)
    df13=clean_transform(xlsx13,csv13)

    #单元测验
    df=[df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12,df13]
    df1=df1[['项目组', '项目组成员','BIP','工作工时']]
    for d in df:
        try:
            d=d[['项目组', '项目组成员','BIP','工作工时']]
            df1=pd.concat([df1,d],axis=0)
        except:
            print(d,' has some bug')
    project_sum=sum_processing(df1)
    project_sum.rename(columns={'工作工时_x':work_h,'日组均工时':work_avgh},inplace=True)
    project_sum=project_sum[[work_avgh]]
    project_sum.to_csv(csvname)

    
```

```python
#周统计汇总
a=pd.read_csv("sum20200620.csv")
b=pd.read_csv("sum20200621.csv")
c=pd.read_csv("sum20200622.csv")
d=pd.read_csv("sum20200623.csv")
e=pd.read_csv("sum20200624.csv")
f=pd.read_csv("sum20200625.csv")
g=pd.read_csv("sum20200626.csv")
h=pd.read_csv("sum20200627.csv")
i=pd.read_csv("sum20200628.csv")

aa=pd.merge(a,b,how='left',on='项目组')
bb=pd.merge(aa,c,how='left',on='项目组')
cc=pd.merge(bb,d,how='left',on='项目组')
dd=pd.merge(cc,e,how='left',on='项目组')
ee=pd.merge(dd,f,how='left',on='项目组')
ff=pd.merge(ee,g,how='left',on='项目组')
gg=pd.merge(ff,h,how='left',on='项目组')
hh=pd.merge(gg,i,how='left',on='项目组')

hh.drop(0,inplace=True)
hh['周均工时']=hh[hh.columns.tolist()[1:10]].sum(axis=1)
hh.to_csv("20206week4-until0628.csv")
```
