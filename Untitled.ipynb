```python
import openpyxl
import pandas as pd
import numpy as np

#filename=file_list[1]

def get_sub_df(filename):
    df=pd.read_excel(filename)
    title=df.columns.tolist()[0]
    index1=np.where(df[[title]]=='项目进展')[0][0]
    index2=np.where(df[[title]]=='项目成员')[0][0]

    #slice into 3 forms
    df_info=df.iloc[0:index1]
    df_wbs=df.iloc[index1:index2]
    df_team=df.iloc[index2:df.shape[0]]

    #forming 3 dictionary

    #1#
    info_dict={}
    df_info=df_info.fillna('')
    for index,row in df_info.iterrows():
        info_dict[row[0]]=row[1]
        info_dict[row[3]]=row[4]

    #2#
    df_wbs[title]=df_wbs[title].fillna(method='ffill')
    df_wbs=df_wbs.fillna('')
    wbs_dict={}
    for index,row in df_wbs.iterrows():
        #注意index的制定
        if index >7:
            wbs_dict[str(row[0])+str(row[1])]=row[2]

    wbs_dict[df_wbs.iloc[1][3]]=df_wbs.iloc[2][3]
    wbs_dict[df_wbs.iloc[1][4]]=df_wbs.iloc[2][4]
    #print(wbs_dict)

    #3#
    df_team=df_team.fillna('')
    inteam_dict={}
    outteam_dict={}
    for index,row in df_team.iterrows():
        #注意index的制定
        if index>37:
            inteam_dict[row[2]]=[row[1],row[0]]
            outteam_dict[row[3]]=[row[4],row[5]]
    #print(outteam_dict)

    df_info=pd.DataFrame(info_dict,index=[0])
    df_wbs=pd.DataFrame(wbs_dict,index=[0])
    df_inteam=pd.DataFrame(inteam_dict)
    df_outteam=pd.DataFrame(outteam_dict)
    return df_info,df_wbs,df_inteam,df_outteam


def merge_df(file_list):
    info=[]
    wbs=[]
    inteam=[]
    outteam=[]
    for filename in file_list:
        df_info=get_sub_df(filename)[0]
        df_wbs=get_sub_df(filename)[1]
        df_inteam=get_sub_df(filename)[2]
        df_outteam=get_sub_df(filename)[3]
        info.append(df_info)
        wbs.append(df_wbs)
        inteam.append(df_inteam)
        outteam.append(df_outteam)
    return info,wbs,inteam,outteam
        
```

```python
info=merge_df(file_list)[0]
wbs=merge_df(file_list)[1]
inteam=merge_df(file_list)[2]
outteam=merge_df(file_list)[3]

def concat_df(df):
    df1=df[0]
    df.remove(df1)
    for d in df:
        try:
            df1=pd.concat([df1,d],axis=0)
        except:
            print(d,' has some bug')
    return df1
```
